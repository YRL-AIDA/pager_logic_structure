```python
from dotenv import load_dotenv
from pathlib import Path
import os 
load_dotenv()
PATH_ICDAR_2013=Path(os.getenv('PATH_ICDAR_2013'))
ICDAR_2013_PDF = PATH_ICDAR_2013/'pdfs'
if not ICDAR_2013_PDF.exists():
    ICDAR_2013_PDF.mkdir()
book_name_list_file = PATH_ICDAR_2013/"2013_books.txt"
```

```python
import requests

with open(book_name_list_file, 'r') as f:
    for line in f:
        bookid, url = line.split()
        name_file = f"{bookid}.pdf"
        path = ICDAR_2013_PDF/name_file
        if path.exists():
            continue
        r = requests.get(url)
        with open(path, "wb") as d:
            d.write(r.content)
```

```python
import pager
from pager.pager_pipe_line import pdf2region_row

from pager.doc_model import MinerPDFModel
from pager.doc_model import LogicTreeModel
from pager.doc_model.converters import PDF2LogicTree

from typing import Dict

pdf_reader = MinerPDFModel({"page_model": pdf2region_row})
pdf2tree = PDF2LogicTree()
tree = LogicTreeModel()
```

```python
FILES = list(ICDAR_2013_PDF.iterdir())
file_name = FILES[1]

file_path = ICDAR_2013_PDF/file_name
pdf_reader.read_from_file(file_path)
pdf_reader.extract()
pdf2tree.convert(pdf_reader, tree)


doc_tree = tree.to_dict()
```
